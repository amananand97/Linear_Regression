```python
import numpy as np
```

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
```

```python
house_price = [245,312,279,308,199,219,405,324,319,255] #Created a list
size = [1400,1600,1700,1875,1100,1550,2350,2450,1425,1700] 
```

```python
plt.scatter(size,house_price,color = "black") 
plt.ylabel("house price") #alloting house price to y-axis.
plt.xlabel("size of house") #alloting size of house to x-axis.
plt.show() #to display/show the graph 
```

```python
size2 = np.array(size).reshape((-1,1)) #created or converted thr list into array
```

```python
regr = linear_model.LinearRegression() #called function "linear_model.LinearRegression()"
```

```python
regr.fit(size2,house_price) #passed the parameters into the regr function
```

```python
def graph(formula, x_range):
    x = np.array(x_range)
    y = eval(formula)
    plt.scatter(size,house_price,color = "black")
    plt.ylabel("house price")
    plt.xlabel("size of house")
    plt.plot(x,y)
```

```python
print("Coefficents = ",regr.coef_)
print("Intercept = ",regr.intercept_)
```

```python
graph("regr.coef_*x + regr.intercept_",range(1000,2700))
```
